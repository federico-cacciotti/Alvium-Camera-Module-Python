```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
import os
from tqdm import tqdm

%matplotlib qt

path = "path/to/folder"
filenames = sorted(glob(f"{path}/*.raw"))

# create output directory if it doesn't exist
output_dir = f"{path}/png_images"
os.makedirs(output_dir, exist_ok=True)

camera_size = (3008, 4128)

# open the file and read the raw data
for filename in tqdm(filenames):
    with open(f"{filename}", "rb") as f:
        raw_data = f.read()

    # open the raw image and resize
    img = np.frombuffer(raw_data, dtype=np.uint8).reshape((camera_size[0], camera_size[1], 3))

    # change image from BGR to RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # save the image as png
    #output_filename = os.path.join(output_dir, os.path.basename(filename).replace('.raw', '.png'))
    #cv2.imwrite(output_filename, img)

plt.imshow(img, cmap='gray')
plt.show()
```
